```python
data = pd.read_excel("00_número operadoras ativas.xlsx")

data.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.style.use('seaborn-darkgrid')

plt.plot(data['Ano'], data['Total de operadoras com beneficiários'], marker='o', color='b', label='Total de operadoras com beneficiários')

plt.title('Total de Operadoras Ativas por Ano', fontsize=15)
plt.xlabel('Ano', fontsize=13)
plt.ylabel('Total de Operadoras', fontsize=13)
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.style.use('seaborn-darkgrid')

palette = plt.get_cmap('Set1')

plt.plot(data['Ano'], data['Total de operadoras com beneficiários'], marker='', color=palette(1), linewidth=2.5, alpha=0.9, label='Total de operadoras com beneficiários')
plt.fill_between(data['Ano'], data['Total de operadoras com beneficiários'], color=palette(1), alpha=0.1)

plt.title('Total de Operadoras com Ativas por Ano', fontsize=24, fontweight='bold', color=palette(1))
plt.xlabel('Ano', fontsize=20, fontweight='bold', color=palette(1))

plt.xticks(fontsize=16)
plt.yticks([])

plt.grid(True, which='major', linestyle='--', linewidth=0.5)
plt.grid(True, which='minor', linestyle='--', linewidth=0.1)

for i, v in enumerate(data['Total de operadoras com beneficiários']):
    plt.text(data['Ano'][i], v + 15, str(v), ha='center', va='bottom', fontsize=16, color=palette(1))



plt.gca().spines['top'].set_alpha(0.7)
plt.gca().spines['bottom'].set_alpha(0.7)
plt.gca().spines['right'].set_alpha(0.7)
plt.gca().spines['left'].set_alpha(0.7)

plt.tight_layout()
plt.show()
```
